```python
import numpy as mp
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
```

```python
dataset = pd.read_csv("dataset.csv")
```

```python
dataset.head()
```

```python
dataset.shape
```

```python
dataset.describe()
```

```python
dataset.info()
```

```python
X = dataset.iloc[:,:-1].values
Y = dataset.iloc[:,-1].values
```

```python
print(X)
```

```python
print(Y)
```

```python
linear_model = LinearRegression()
linear_model.fit(X, Y)
```

```python
linear_pred = linear_model.predict(X)
```

```python
plt.scatter(X, Y, color='blue')
plt.plot(X, linear_pred, color='red')
plt.xlabel('Level')
plt.ylabel('Salary')
plt.title('Prediction using Linear Regression')
plt.tight_layout()
plt.show()
```

```python
poly_featuring = PolynomialFeatures(degree=4)
x_poly = poly_featuring.fit_transform(X)
```

```python
model_poly = LinearRegression()
model_poly.fit(x_poly, Y)
```

```python
y_pred = model_poly.predict(x_poly)
```

```python
plt.scatter(X, Y, color='blue')
plt.plot(X, y_pred, color='red')
plt.show()
```

```python
test = 8.5
x_transform = poly_featuring.fit_transform([[test]])
y_pred = model_poly.predict(x_transform)
print(y_pred)
```
