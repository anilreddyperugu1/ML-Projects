**Importing dependencies**

```python
import numpy as mp
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
```

**---- DATA ANALYSIS ---**

```python
dataset = pd.read_csv("dataset.csv") #reading the dataset
```

```python
dataset.head() #gives top 5 rows
```

```python
dataset.shape #gives dimensions of the dataset
```

```python
dataset.describe() #gives statistical values of the data
```

```python
dataset.info() #information about the dataset
```

**Dividing the data into X and Y**

```python
X = dataset[['Level']] #independent column into X
Y = dataset[['Salary']] #dependent column into Y
```

**Analysis of the above operation**

```python
print(X)
```

```python
print(Y)
```

**Intializing Linear Model**

```python
linear_model = LinearRegression()
linear_model.fit(X, Y) #training linear model
```

**Prediction with Linear model**

```python
linear_pred = linear_model.predict(X)
```

**---- VISUALIZATION OF LINEAR MODEL PREDICTION ----**

```python
plt.scatter(X, Y, color='blue') #scatter plot
plt.plot(X, linear_pred, color='red') #plotting the line
plt.xlabel('Level') #xlabel
plt.ylabel('Salary') #ylabel
plt.title('Prediction using Linear Regression') #graph title
plt.tight_layout()
plt.show()
```

**Intializing Polynomial Model**

```python
poly_featuring = PolynomialFeatures(degree=4)
x_poly = poly_featuring.fit_transform(X) #transforming the features w.r.t the degree
```

**Fitting the data**

```python
model_poly = LinearRegression()
model_poly.fit(x_poly, Y) #training the model
```

```python
y_pred = model_poly.predict(x_poly)
```

**---- VISUALIZATION OF POLYNOMIAL MODEL PREDICTION ----**

```python
plt.scatter(X, Y, color='blue')
plt.plot(X, y_pred, color='red')
plt.xlabel('Level') #xlabel
plt.ylabel('Salary') #ylabel
plt.title('Prediction using Polynomial Regression') #graph title
plt.tight_layout()
plt.show()
```

**---- MANUAL PREDICTION ----**

```python
test = 8.5
x_transform = poly_featuring.fit_transform([[test]])
y_pred = model_poly.predict(x_transform)
print(y_pred)
```
