# **Car Price Prediction using RANDOM FOREST**

**Importing the dependencies**

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score, accuracy_score
```

**---- DATA ANALYSIS & PRE PROCESSING ----**

```python
dataset = pd.read_csv('dataset.csv') #reading the dataset
```

```python
dataset.head() #gives 1st 5 rows
```

```python
dataset.tail() #gives last 5 rows
```

```python
dataset.shape #dimensions of the dataset
```

**Dropping the useless columns**

```python
dataset = dataset.drop(columns='car_ID', index=1) 
```

**Analysis of above operation**

```python
dataset.head() #column dropped
```

**Checking if any null values**

```python
dataset.isnull().sum() #No null values present
```

```python
dataset['CarName'].value_counts()
```

**Dividing data into X and Y for encoding & splitting**

```python
X = dataset.iloc[:, :-1] #independent columns into X 
Y = dataset['price'] #dependent column into Y
```

**Analysis of the above division**

```python
print(X)
```

```python
X.info()
```

```python
print(Y)
```

```python
X.shape
```

```python
Y.shape
```

```python
print(Y)
```

**ONE HOT ENCODING the data**

```python

categorical_cols = ['CarName', 'fueltype', 'aspiration', 'doornumber', 'carbody', 'drivewheel', 'enginelocation', 'enginetype', 'cylindernumber', 'fuelsystem', ]

X_encoded = pd.get_dummies(X, columns=categorical_cols, drop_first=False) #one hot encoded for the above features
```

```python
X_encoded.head()
```

```python
X_encoded.shape
```

**----- BUILDING A MODEL -----**

**Splitting the data for training and testing**

```python
X_train, X_test, Y_train, Y_test  =train_test_split(X_encoded, Y, test_size=0.2, random_state=87) #Splitting the data(80% for train, 20% for test)
```

**Analysis of above split**

```python
print(X_train.shape) #dimensions of the splitted data
print(X_test.shape) 
```

```python
print(Y_train.shape) #dimensions of the splitted data
print(Y_test.shape)
```

**Initializing a model and training it**

```python
model = RandomForestRegressor(n_estimators=200, random_state=54)
model.fit(X_train, Y_train) #training the model
```

**Prediction on test data**

```python
y_pred = model.predict(X_test) #prediction
print(y_pred)
```

**----- MODEL EVALUATION ----**

**Evaluation Metrics**

```python
MAE = mean_absolute_error(Y_test, y_pred)
MSE = mean_squared_error(Y_test, y_pred)
RMSE = np.sqrt(MSE)
R2_score = r2_score(Y_test, y_pred)

print('MAE :', MAE)
print('MSE :', MSE)
print('RMSE :', RMSE)
print('R2 score :', R2_score)
```
