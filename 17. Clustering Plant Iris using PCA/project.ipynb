# **Clustering Plant Iris Using Principal component Analysis**

**Importing dependencies**

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import load_iris
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

**---- INITIAL DATA ANALYSIS & PRE PROCESSING ----**

```python
dataset = load_iris() #loading the dataset
```

```python
dataset.data #data
```

```python
dataset.data.shape #dimensions of the data
```

**Dividing into X and Y**

```python
X = pd.DataFrame(dataset.data, columns=dataset.feature_names) #Independent features
Y = pd.Series(dataset.target, name="species") #dependent feature
```

**Analysis of the above opearation**

```python
print(X.head())
```

```python
print(Y.unique())  # Target classes 
```

**SCALING the data**

```python
SC = StandardScaler()
X_transformed = SC.fit_transform(X) #scaling the data
```

```python
print(X_transformed)
```

**Intializing PCA and training it**

```python
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X_transformed) #training & transforming the data
```

**Dimensionality reduced data**

```python
print(X_pca)
```

**---- PCA MODEL EVALUATION ----**

```python
print("Explained Variance Ratio:", pca.explained_variance_ratio_)
print("Total variance retained:", sum(pca.explained_variance_ratio_))
```

**---- VISUALIZATION ----**

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(8,6))
plt.scatter(X_pca[:,0], X_pca[:,1], c=Y, cmap='viridis', edgecolor='k', s=80) #plotting the data
plt.xlabel("Principal Component 1")
plt.ylabel("Principal Component 2")
plt.title("PCA on Iris Dataset")
plt.colorbar(label="Species")
plt.show()
```

**---- MAIN MODEL BUILDING ----**

```python
X_train, X_test, y_train, y_test = train_test_split(X_pca, Y, test_size=0.2, random_state=42) #Splitting data for train and test
```

**Intialized the model**

```python
model = LogisticRegression()
model.fit(X_train, y_train) #training the model
```

**Finding actual model's accuracy**

```python
y_pred = model.predict(X_test)
print("Accuracy:", accuracy_score(y_test, y_pred)) #accuracy
```
